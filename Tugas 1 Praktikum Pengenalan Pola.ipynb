```python
import numpy as sp #import library numpy
traffic = sp.genfromtxt("web_traffic.tsv",delimiter='\t') #membaca data "web_traffic.tsv" yang sudah dimasukkan ke dalam server
print(traffic[:10]) #mencetak data traffic dari 1 sd 10
print(traffic.shape) #array merupakan dim traffic dengan jumlah elemen shape

x = traffic[:,0] #data dari awal sd ,0
y = traffic[:,1] #data dari awal sd ,1

x = x[~sp.isnan(y)]
y = y[~sp.isnan(y)]

import matplotlib.pyplot as plt #import library matplotlib
plt.scatter(x,y) #bentuk grafik scatterplot
plt.title("Web traffic last month") #judul grafik
plt.xlabel("Time") #nama label x
plt.ylabel("Hits/hour") #nama label y

plt.xticks([w*7*24 for w in range(10)],['week %i' %w for w in range(10)]) #nama elemen di label x
plt.autoscale(tight=True)
plt.grid()
```

```python
# Operasi image pada Python 
# pilihan load image (contoh logo ipb)
import matplotlib.pyplot as plt #import library matplotlib
import cv2 #import library cv2
import numpy as np #import library numpy

print("read images using opencv") #mencetak kalimat "read images using opencv"
five = cv2.imread("5.png") #memproses gambar logo IPB (5) yang sudah dimasukkan ke dalam server
print(five.shape) #mencetak batas awal, batas akhir, dengan jumlah setiap elemen 3
print(five.size) #mencetak ukuran gambar
plt.imshow(five) #mencetak gambar
cv2.waitKey(0)
```
